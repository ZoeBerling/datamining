```python
# Part 1

import numpy as np
import pandas as pd

df = pd.read_csv("exercise8_infile1.csv")

def distance(p1, p2):
    p1x = p1[0]
    p1y = p1[1]
    p1z = p1[2]
    p2x = p2[0]
    p2y = p2[1]
    p2z = p2[2]
    distval = np.sqrt((p1x - p2x) ** 2 + (p1y - p2y) ** 2 + (p1z - p2z) ** 2)
    return distval

def centroid(cluster):
    xsum = 0
    ysum = 0
    zsum = 0
    i = 0
    for point in cluster:
        xsum += point[0]
        ysum += point[1]
        zsum += point[2]
        i += 1
    xmean = xsum/i
    ymean = ysum/i
    zmean = zsum/i
    return([xmean, ymean, zmean])

centr1 = df.iloc[np.random.randint(0, len(df))].to_list()
centr2 = df.iloc[np.random.randint(0, len(df))].to_list()
centr3 = df.iloc[np.random.randint(0, len(df))].to_list()

tempcentr1 = [-1, -1, -1]
tempcentr2 = [-1, -1, -1]
tempcentr3 = [-1, -1, -1]

while(centr1 != tempcentr1) and (centr2 != tempcentr2) and (centr3 != tempcentr3):
    clust1 = []
    clust2 = []
    clust3 = []
    for index, row in df.iterrows():
        dist1 = distance(row, centr1)
        dist2 = distance(row, centr2)
        dist3 = distance(row, centr3)
        if dist1 == min(dist1, dist2, dist3):
            clust1.append(row.to_list())
        if dist2 == min(dist1, dist2, dist3):
            clust2.append(row.to_list())
        if dist3 == min(dist1, dist2, dist3):
            clust3.append(row.to_list())
    tempcentr1 = centr1
    tempcentr2 = centr2
    tempcentr3 = centr3
    centr1 = centroid(clust1)
    centr2 = centroid(clust2)
    centr3 = centroid(clust3)

textfile = open("handCodedOut.txt", 'w')
textfile.write("Cluster 1: " + str(len(clust1)) + " values\n\n")
for row in clust1:
    textfile.write(str(row) + '\n')
textfile.write("Cluster 2: " + str(len(clust2)) + " values\n\n")
for row in clust2:
    textfile.write(str(row) + '\n')
textfile.write("Cluster 3: " + str(len(clust3)) + " values\n\n")
for row in clust3:
    textfile.write(str(row) + '\n')
textfile.close()
```
